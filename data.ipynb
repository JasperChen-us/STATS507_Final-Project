```python
import yfinance as yf
import pandas as pd

# 下载 BTC-USD 历史数据
df = yf.download(
    "BTC-USD",
    start="2020-01-01",
    end="2026-04-13",
    interval="1d",
    auto_adjust=False,   # 保留原始 OHLC
    progress=False
)

# 整理列名
df = df.rename(columns={
    "Open": "open",
    "High": "high",
    "Low": "low",
    "Close": "close",
    "Volume": "volume"
})

print(df.head())

# 保存
df.to_csv("btc_data.csv", index=True)
```
